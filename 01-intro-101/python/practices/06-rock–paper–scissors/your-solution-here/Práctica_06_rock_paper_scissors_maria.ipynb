## 1.1. Ejercicio sin Bonus

```python
# Importamos la librería random.
import random


# Creamos la lista con las opciones válidas de juego y algunas variables.
list_options = ["piedra", "papel", "tijeras"]
maximum_games = 3
num_games_to_win = 2


# Definimos una función para obtener la respuesta elegida por la máquina.
def get_random_choice(options):
    return random.choice(options)


# Definimos una función que pregunte la elección del usuario.
def get_user_choice():
    user_choice = input("Introduce una de las siguientes opciones: piedra, papel o tijeras")
    while user_choice not in list_options:
        print("No has elegido una de las opciones disponibles, por favor, intentalo de nuevo")
        user_choice = input("Introduce una de las siguientes opciones: piedra, papel o tijeras")
    return user_choice


# Definimos una función que resuelva un combate, devolviendo distintos valores en función de si gana la máquina, gana el
# usuario o hay empate.
def fight(random_choice, user_choice):
    if random_choice == "tijeras" and user_choice == "tijeras":
        return 0
    elif random_choice == "piedra" and user_choice == "piedra":
        return 0
    elif random_choice == "papel" and user_choice == "papel":
        return 0
    elif random_choice == "piedra" and user_choice == "papel":
        return 2
    elif random_choice == "piedra" and user_choice == "tijeras":
        return 1
    elif random_choice == "papel" and user_choice == "piedra":
        return 1
    elif random_choice == "papel" and user_choice == "tijeras":
        return 2
    elif random_choice == "tijeras" and user_choice == "piedra":
        return 2
    elif random_choice == "tijeras" and user_choice == "papel":
        return 1

    
# Definimos una función que muestre por pantalla la elección de la máquina y el jugador y el estado de la partida.
def print_state(machine, user, win_machine, win_user):
    print("La maquina ha elegido " + machine)
    print("El jugador ha elegido " + user)
    print("La maquina lleva ganadas " + str(win_machine) + " partidas")
    print("El jugador lleva ganadas " + str(win_user) + " partidas")

    
# Creamos dos variables que acumulen las partidas ganadas por la máquina y el jugador.
machine_counter = 0
user_counter = 0


# Creamos un bucle que itere mientras nadie alcance el mínimo de victorias necesarias para ganar, previamente establecidas.
# Dentro del bucle, resolvemos la jugada y comparamos y actualizamoso el valor de las variables que acumulen las partidas 
# ganadas de la máquina y el jugador.
while machine_counter < num_games_to_win and user_counter < num_games_to_win:
    m = get_random_choice(list_options)
    u = get_user_choice()
    result = fight(m, u)
    if result == 1:
        machine_counter += 1
        print("Ha ganado la máquina")
    elif result == 2:
        user_counter += 1
        print("Ha ganado el jugador")
    else:
        print("Empate")
    print_state(m, u, machine_counter, user_counter)

    
# Mostramos por pantalla el ganador del juego en función de quién tiene más victorias aculumadas.
if user_counter > machine_counter:
    print("El jugador ha acumulado más victorias")
elif user_counter < machine_counter:
    print("La máquina ha acumulado más victorias")
else:
    print("Ha habido un empate")
```
