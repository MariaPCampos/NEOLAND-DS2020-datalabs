## Descripción: 
En el espacio de 2 dimensiones un punto se puede definir por un par de valores que corresponden a la coordenada horizontal (x) y a la vertical (y). El espacio puede quedar dividido en 4 zonas (cuadrantes): Q1, Q2, Q3, Q4. Cuyo punto de unión único es el punto (0, 0). 

Si un punto se encuentra en Q1 tanto su coordenada x como la y son positivas. Te dejo un enlace a wikipedia para que te familiarices con estos cuadrantes.

## Tareas
1. Robin Hood es famoso por acertar a una flecha con otra flecha. ¿Lo ha conseguido? 
2. Calcula cuántos flechazos han caido en cada cuadrante. 
3. Halla el punto más cercano al centro. Calcula su distancia al centro. Definir una función que calcula la distancia al centro puede servir de ayuda.
4. Si la diana tiene un radio de 9, calcula el número de flechas que hay que recoger al bosque. 

```python
points = [(4,5), (-0,2), (4,7), (1,-3), (3,-2), (4,5), 
          (3,2), (5,7), (-5,7), (2,2), (-4,5), (0,-2), 
          (-4,7), (-1,3), (-3,2), (-4,-5), (-3,2), 
          (5,7), (5,7), (2,2), (9, 9), (-8,-9)]

# Creo una lista vacia donde voy introduciendo los valores al recorrer la lista "points" mientras cumplan la condicion de
# no ser valores repetidos. En el momento que encuentre el valor repetido en la lista sale del bucle, Robin ha acertado.

new_values = []

for i in points:
    if new_values.count(i) == 0:
        new_values.append(i)
    else:
        print("Robin Hood hit an arrow with another one")
        break

        
# Calculos de cuantos flechazos han caido en cada cuadrante.

quadrant_one = []
quadrant_two = []
quadrant_three = []
quadrant_four = []

for i in points:
    if i[0] > 0 and i[1] > 0:
        quadrant_one.append(i)
    elif i[0] < 0 and i[1] > 0:
        quadrant_two.append(i)
    elif i[0] < 0 and i[1] < 0:
        quadrant_three.append(i)
    elif i[0] > 0 and i[1] < 0:
        quadrant_four.append(i)

print(str(len(quadrant_one)) + " arrows have hit in quadrant one")
print(str(len(quadrant_two)) + " arrows have hit in quadrant two")
print(str(len(quadrant_three)) + " arrows have hit in quadrant three")
print(str(len(quadrant_four)) + " arrows have hit in quadrant four")


# Creamos una funcion para calcular la distancia al centro desde cualquier punto de la diana.

import math

def center_distance(p):
    return math.sqrt(p[0]**2 + p[1]**2)


# Hallamos el punto mas cercano al centro haciendo uso de la funcion.
minimum_point = points[0]


for i in points:
    if center_distance(minimum_point) > center_distance(i):
        minimum_point = i
        
print("The closest point to the center is " + str(minimum_point))

# Calculamos el numero de flechas que quedan fuera de la diana, considerando la distancia al centro en base al radio de la
# circunferencia. Si la distancia al centro de los puntos esta por encima de nueve, los incluimos en la lista que guardara
# los puntos que han quedado fuera de la diana.

arrows_in_the_forest = []

for i in points:
    if center_distance(i) > 9:
        arrows_in_the_forest.append(i)
print("The number of arrows in the forest is " + str(len(arrows_in_the_forest)))
```
